#### 3. В центре 95% доверительного интервала, рассчитанного по выборочным значениям, находится:

    1) Значение стандартной ошибки среднего

Доверительный интервал представляет собой диапазон значений, в котором находится неизвестный параметр с определенной вероятностью. Центр доверительного интервала, как правило, находится вблизи выборочного среднего или выборочной медианы, поскольку они являются статистиками, которые обычно используются для оценки параметров.

Центр доверительного интервала может меняться в зависимости от уровня доверия, выбранного при рассчете. В нашем случаи при доверительном интервала с 95% уровнем доверия, центр интервала будет находиться на значении, которое включает в себя 95% всех возможных значений.

**Важно**, что центр доверительного интервала является оценкой и может отличаться от истинного значения параметра из-за случайности выборки. Поэтому **доверительный интервал предоставляет информацию о диапазоне значений, в котором находится истинный параметр с заданной вероятностью, а не точное значение параметра.**

#### 4. Часто на практике нулевая гипотеза отклоняется, и различия считаются статистически достоверными, если p < 0,05. Однако часто в статистике используется более жесткий критерий достоверности различий, например, при условии, что p < 0,01. Значение p-уровня значимости, которое выбирается, в качестве порога обозначается буквой α (альфа). Например, если исследователь решил, что α = 0,05, то и нулевая гипотеза будет отклоняться при условии, что p < 0,05. 

#### Если в определенной ситуации весьма рискованно отклонить нулевую гипотезу, когда она на самом деле верна, то лучше использовать показатель α равный 

Выберите один вариант из списка

    1) 0,1
    2) 0,001
    3) 0,05
    4) 0,5

Как было сказанно выше, обычно уровень значимости α устанавливают на уровне 0.05 или 0.01, что соответствует 5% или 1% рискам ошибки первого рода. Ошибка первого рода возникает, когда мы отклоняем нулевую гипотезу, хотя она на самом деле верна.

В ситуациях, когда риски ошибки первого рода являются критическими или нежелательными, можно рассмотреть использование более низкого уровня значимости, например, 0.01. Это позволяет уменьшить вероятность отклонения нулевой гипотезы при условии ее истинности. Исходя из этого я бы взяла 0.01, но такого варианта тут нет, что плохо и как я понимаю правильным будет вариант 0.001, но на мой взгляд он очень низкий, т.к.  следует помнить, что уменьшение уровня значимости ведет к увеличению риска ошибки второго рода, которая происходит, когда мы принимаем нулевую гипотезу, когда она на самом деле не верна. Поэтому необходимо балансировать риски ошибок первого и второго рода в контексте конкретной статистической ситуации и принять соответствующее решение об уровне значимости.

    3) Доверительный интервал [160, 172] также не включает значение 158, поэтому у нас есть основания отклонить нулевую гипотезу, что среднее в генеральной совокупности равняется 158.

    4) Данные о выборочном стандартном отклонении не дают основания для вывода о ширине доверительного интервала. Размер доверительного интервала зависит от выбранного уровня доверия и дисперсии в генеральной совокупности, а не только от выборочного стандартного отклонения..



#### 6. Предположим, нулевой гипотезой вашего исследования являлось предположение, что конверсия в генеральной совокупности равняется 0.4. Вы получили p = 0,12 и не смогли отклонить нулевую гипотезу. Однако позже выяснилось, что конверсия в генеральной совокупности действиетльно равна 0.4. Как можно оценить результаты?


Какой вариант корректный по вашему мнению и почему?

    1) Вы не совершали ни ошибку первого рода, ни ошибку второго рода.
    2) Вы совершили ошибку первого рода
    3) Вы совершили ошибку второго рода

Так как мы не смогли отклонить нулевую гипотезу о том, что конверсия в генеральной совокупности равна 0.4 на основе полученного значения p=0.12, и позже выяснили, что конверсия действительно равна 0.4, то мы  **не совершали ни ошибку первого рода, ни ошибку второго рода.**

Ошибка первого рода (False Positive) была бы, если бы мы отклонили нулевую гипотезу, чего мы не сделали.

Ошибка второго рода (False Negative) произошла бы, если бы мы приняли нулевую гипотезу, а она не верна, чего тоже не произошло. Следовательно мы не совершили ни ошибку первого ни ошибку второго рода.
